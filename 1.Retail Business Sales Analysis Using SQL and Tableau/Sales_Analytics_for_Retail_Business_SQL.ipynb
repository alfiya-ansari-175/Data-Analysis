# Depth Analysis

```python
%%sql
select * from online_retail limit 4;
```

## 1. Total Sales Revenue

```python
%%sql
select sum(Quantity * UnitPrice) as TotalSales from online_retail;
```

## 2.Average Order Value

```python
%%sql
Select AVG(invoice_total) from (
    select InvoiceNo, SUM(Quantity * UnitPrice) AS invoice_total
        from online_retail group by InvoiceNo
)As invoice_totals;
```

## 3. Top 10 Customers by Total Spending

```python
%%sql
select * from online_retail limit 2;
```

```python
%%sql
select CustomerID,SUM(Quantity * UnitPrice) AS total_spent from online_retail
        group by CustomerID order by total_spent DESC Limit 10;
```

## 4.Customer Purchase Frequently

```python
%%sql
select CustomerID, count(distinct InvoiceNo) as purchase_count from online_retail group by CustomerID 
            order by purchase_count desc limit 5;
```

## 5.Recently (last) Purchase Date Per Customer

```python
%%sql
select CustomerID,MAX(InvoiceDate) as last_purchase from online_retail group by CustomerID limit 5;
```

## 6.Customer Segmentation by Country

```python
%%sql
select Country, count(distinct CustomerID) as customer_count from online_retail group by Country order by customer_count desc limit 5;
```

# 7.Count of One Time Buyers

```python
%%sql
WITH one_time_customers AS(
    select CustomerID from online_retail group by CustomerID HAVING COUNT(DISTINCT InvoiceNo) = 1
)
select count(*) As OneTimeBuyers from one_time_customers
```

# 8. Revenue By Conutry

```python
%%sql
Select Country, sum(Quantity*UnitPrice) as revenue from online_retail group by Country order by revenue desc limit 3
```

# 9.Average Revenue Per Invoice

```python
%%sql
WITH avg_rev_per_invoice AS(
    select InvoiceNo, SUM(Quantity*UnitPrice) as invoice_total from online_retail group by InvoiceNo
)
select avg(invoice_total) from avg_rev_per_invoice
```

# 10.Revenue per StockCode

```python
%%sql
select StockCode, SUM(Quantity*UnitPrice) as Revenue from online_retail group by StockCode order by Revenue desc limit 2;
```

# 11.Top Selling Items

```python
%%sql
select Description, SUM(Quantity) as total_quantity from online_retail group by Description 
                    order by total_quantity desc limit 5;
```

# 12.Top Revenue Generating Items

```python
%%sql
select Description, SUM(Quantity * UnitPrice) as revenue from online_retail group by Description 
                    order by revenue desc limit 5;
```

# 13.Number of Unique Products

```python
%%sql
select count(distinct StockCode) As unique_products from online_retail;
```

# 14.Products Sold in Most Countries

```python
%%sql
select Description, count(distinct Country) as country_count from online_retail group by Description 
            order by country_count desc limit 5;
```

# 15.Products with Missing/Unclear Description

```python
%%sql
select * from online_retail where Description is null or Description='';
```

# 16.Low Selling Products

```python
%%sql
select Description, SUM(Quantity) as total_quantity from online_retail group by Description HAVING total_quantity<10
                order by total_quantity limit 10;
```

# 17.Christmas Season Sales

```python
%%sql
select * from online_retail where MONTH(InvoiceDate) = 12;
```

# 18.Revenue By Country

# 19.Top Products Per Country
